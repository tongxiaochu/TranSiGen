## Fig 2D

### (1)相同作用机制和baseline均采样sample count数量的谱计算pearson, 并且统计数量大于sample count的作用机制

```python
# cell_ls = ['A375', 'HA1E', 'HELA', 'HT29', 'MCF7', 'PC3', 'YAPC']
## sample count 从1开始统计
sample_count = 1
random_seed = 5678
df_moa = pd.DataFrame(columns=['moa', 'pearson', 'pearson_type', 'cid', 'data'])
for selected_cid in cell_ls:
    # selected_cid = 'PC3'
    df_LINCS_cid = df_LINCS[(df_LINCS['cid']==selected_cid) & (df_LINCS['moa']!='none')]
    for moa, group in df_LINCS_cid.groupby('moa'):
        # print(moa)
        # print(group)
        index_ls = group.index
        # print(index_ls)
        if group.shape[0]> sample_count:

            pearson_inner_ls = []
            pearson_outer_ls = []
            #for data_name, data_array in zip(['DEGs$_{real}$', 'DEGs$_{rec}$', 'SigGen'], 
            #                                   [DEG_array,  DEG_rec_array, DEG_pred_array]):
            for data_name, data_array in zip(['$\mathit{\Delta}X$', 'TranSiGen'], 
                                               [DEG_array,  DEG_pred_array]):
                for idx in index_ls:
                    # print(idx)
                    # df_LINCS_inner = group[~group.index.isin([idx])]
                    # df_LINCS_outer = df_LINCS_cid[~df_LINCS_cid.index.isin(index_ls)]
                    
                    df_LINCS_inner = group[~group.index.isin([idx])].sample(n=sample_count, random_state=random_seed)
                    df_LINCS_outer = df_LINCS_cid[~df_LINCS_cid.index.isin(index_ls)].sample(n=sample_count, random_state=random_seed)

                    DEG = data_array[idx,:]
                    DEG_inner = data_array[df_LINCS_inner.index,:]
                    DEG_outer = data_array[df_LINCS_outer.index,:]
                    for idy in range(DEG_inner.shape[0]):
                        # print(DEG.shape, DEG_inner[idy,:].shape)
                        pearson_inner = pearsonr(DEG, DEG_inner[idy,:])[0]
                        pearson_inner_ls.append(pearson_inner)
                    for idy in range(DEG_outer.shape[0]):
                        # print(DEG.shape, DEG_inner[idy,:].shape)
                        pearson_outer = pearsonr(DEG, DEG_outer[idy,:])[0]
                        pearson_outer_ls.append(pearson_outer)
               #  print(len(pearson_inner_ls), len(pearson_outer_ls))
                df_moa.loc[df_moa.shape[0],:] = [moa, np.mean(pearson_inner_ls), 'same MOA', selected_cid, data_name]
                df_moa.loc[df_moa.shape[0],:] = [moa, np.mean(pearson_outer_ls), 'different MOA', selected_cid, data_name]
df_moa
```
