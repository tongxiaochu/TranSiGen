```python
df_relation = pd.DataFrame(columns=['data', 'cid', 'value', 'relation_type', 'target'])
target_ls = df_target_for_cid_copy.head(10)['target'].tolist()

# cell_ls = ['A375', 'HA1E', 'HELA', 'HT29', 'MCF7', 'PC3', 'YAPC']

random_seed = 961104
negative_count = 1   #5
negative_random_seed = 961104

for selected_cid in cell_ls:
    LINCS_smiles_ls_for_cid = sorted(list(set(df_LINCS[df_LINCS['cid'] == selected_cid]['canonical_smiles'])))  
    ## smiles for selected_cid
    df_LINCS_smiles_ls_for_cid = pd.DataFrame(LINCS_smiles_ls_for_cid, columns=['canonical_smiles'])
    target_df_for_cid = target_df[(target_df['canonical_smiles'].isin(LINCS_smiles_ls_for_cid))]
    for target in target_ls:
        target_smi_ls = target_df_for_cid[target_df_for_cid['target'] == target]['canonical_smiles'].tolist()

        target_smi_for_cid_target_all = target_df[(target_df['target'] == target)]['canonical_smiles'].tolist()
        print('all smiles for target:', len(target_smi_for_cid_target_all))
        ## 在该细胞系的所有分子中，选择不属于7个细胞系的公共分子以外的分子, 以及该细胞系下，该靶点的其他阳性分子
        untarget_smi_ls = df_LINCS_smiles_ls_for_cid[~df_LINCS_smiles_ls_for_cid['canonical_smiles'].isin(
            set(LINCS_smiles_ls_for_cid_common) | set(target_smi_for_cid_target_all)
        )].sample(n=len(target_smi_ls) * negative_count, replace=False, random_state=negative_random_seed)[
            'canonical_smiles'].tolist()
        ## 在该细胞系的所有分子中，选择该靶标阳性以外的其他分子
        # untarget_smi_ls = df_LINCS_smiles_ls_for_cid[~df_LINCS_smiles_ls_for_cid['canonical_smiles'].isin(target_smi_ls)].sample(n=len(target_smi_ls)*negative_count, replace=False, random_state=0)['canonical_smiles'].tolist()

        print('target smiles count:', len(target_smi_ls), 'untarget smiles count:', len(untarget_smi_ls),
              'common smiles:', set(target_smi_ls) & set(untarget_smi_ls))
        print('common smiles for untarget_smi & all active for target:',
              set(target_smi_for_cid_target_all) & set(untarget_smi_ls))

        df_LINCS_for_cid_target = df_LINCS[
            (df_LINCS['cid'] == selected_cid) & (df_LINCS['canonical_smiles'].isin(target_smi_ls))]
        df_LINCS_for_cid_target.loc[:, 'label'] = 1
        df_LINCS_for_cid_untarget = df_LINCS[
            (df_LINCS['cid'] == selected_cid) & (df_LINCS['canonical_smiles'].isin(untarget_smi_ls))]
        df_LINCS_for_cid_untarget.loc[:, 'label'] = 0
        df_LINCS_for_cid_target_all = pd.concat([df_LINCS_for_cid_target, df_LINCS_for_cid_untarget])

        for data_name, data_new in zip(['$\mathit{\Delta}X$', 'TranSiGen',], [DEG_array, DEG_pred_array]):
            print('=================', selected_cid, target, data_name, '====================')
            index_ls = df_LINCS[
                (df_LINCS['cid'] == selected_cid) & (df_LINCS['canonical_smiles'].isin(target_smi_ls))].index
            response = data_new[index_ls, :]

            index_ls = df_LINCS[
                (df_LINCS['cid'] == selected_cid) & (df_LINCS['canonical_smiles'].isin(untarget_smi_ls))].index
            no_response = data_new[index_ls, :]

            pred_response_inter_pearson_ls = []
            for idx in range(response.shape[0]):
                for idy in range(idx + 1, response.shape[0]):
                    pear = pearsonr(response[idx, :], response[idy, :])[0]
                    pred_response_inter_pearson_ls.append(pear)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, pear, 'Active', target]
            print('pred_response_inter_pearson:', len(pred_response_inter_pearson_ls))
            # print('pred_response_inter_pearson:', round(np.mean(pred_response_inter_pearson_ls),3))

            pred_no_response_inter_pearson_ls = []
            for idx in range(no_response.shape[0]):
                for idy in range(idx + 1, no_response.shape[0]):
                    pear = pearsonr(no_response[idx, :], no_response[idy, :])[0]
                    pred_no_response_inter_pearson_ls.append(pear)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, pear, 'Inactive', target]
            # print('pred_no_response_inter_pearson:', round(np.mean(pred_no_response_inter_pearson_ls),3))

            pred_response_no_response_pearson_ls = []
            for idx in range(response.shape[0]):
                for idy in range(no_response.shape[0]):
                    pear = pearsonr(response[idx, :], no_response[idy, :])[0]
                    pred_response_no_response_pearson_ls.append(pear)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, pear, 'Active ~ Inactive', target]
            print('pred_response~no_response_pearson:', len(pred_response_no_response_pearson_ls))
            print(data_name, selected_cid, np.mean(pred_response_inter_pearson_ls),
                  np.mean(pred_no_response_inter_pearson_ls),
                  np.mean(pred_response_no_response_pearson_ls))

        for data_name, data_new in zip(['ECFP4'], [ECFP_array]):
            print('=================', selected_cid, target, data_name, '====================')
            response = []
            response_smiles = target_smi_ls
            for smi in response_smiles:
                mol = Chem.MolFromSmiles(smi)
                ECFP = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
                response.append(ECFP)

            no_response = []
            no_response_smiles = untarget_smi_ls
            for smi in no_response_smiles:
                mol = Chem.MolFromSmiles(smi)
                ECFP = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
                no_response.append(ECFP)

            pred_response_inter_pearson_ls = []
            for idx in range(len(response)):
                for idy in range(idx + 1, len(response)):
                    # pred_response_inter_pearson_ls.append(pearsonr(response[idx, :], response[idy, :])[0])
                    sims = DataStructs.TanimotoSimilarity(response[idx], response[idy])
                    pred_response_inter_pearson_ls.append(sims)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, sims, 'Active', target]
            print('pred_response_inter_pearson:', round(np.mean(pred_response_inter_pearson_ls), 3))

            pred_no_response_inter_pearson_ls = []
            for idx in range(len(no_response)):
                for idy in range(idx + 1, len(no_response)):
                    # pred_no_response_inter_pearson_ls.append(pearsonr(no_response[idx, :], no_response[idy, :])[0])
                    sims = DataStructs.TanimotoSimilarity(no_response[idx], no_response[idy])
                    pred_no_response_inter_pearson_ls.append(sims)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, sims, 'Inactive', target]
            # print('pred_no_response_inter_pearson:', round(np.mean(pred_no_response_inter_pearson_ls),3))

            pred_response_no_response_pearson_ls = []
            for idx in range(len(response)):
                for idy in range(len(no_response)):
                    # pred_response_no_response_pearson_ls.append(pearsonr(response[idx, :], no_response[idy, :])[0])
                    sims = DataStructs.TanimotoSimilarity(response[idx], no_response[idy])
                    pred_response_no_response_pearson_ls.append(sims)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, sims, 'Active ~ Inactive', target]
            print('pred_response~no_response_pearson:', len(pred_response_no_response_pearson_ls))


df_relation.to_csv('../results/4.Ligand_based_virtual_screening/relation_within_actives_and_actives~inactives.csv', index=False)
```

```python
def grid_search_for_RF(data_pred, label, random_state, param_grid, scoring, fold):

    clf = ensemble.RandomForestClassifier(random_state=random_state)
    grid_search = GridSearchCV(clf, param_grid, scoring=scoring, cv=fold, n_jobs=6, verbose=2)
    grid_search.fit(data_pred, label)
    return grid_search.best_score_, grid_search.best_params_
```
