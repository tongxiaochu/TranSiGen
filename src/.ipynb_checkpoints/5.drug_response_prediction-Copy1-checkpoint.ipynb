```python
# def eval_metrics_for_group(df):
#     cid_count = 0
#     smi_count = 0
#     metrics_dict_cid = defaultdict(float)
#     metrics_dict_smi = defaultdict(float)
#     for cid, group in df.groupby('cid'):  #一个细胞系上所有分子统计pearson
#         if group.shape[0]>1:
#             cid_count += 1
#             metrics_dict_cid['Pearson'] += pearsonr(group['AUC'], group['AUC_pred'])[0]
#             metrics_dict_cid['RMSE'] += np.sqrt(mean_squared_error(group['AUC'], group['AUC_pred']))
#             metrics_dict_cid['R2'] += r2_score(group['AUC'], group['AUC_pred'])
#     for smi, group in df.groupby('smiles_idx'):  #一个分子在多个细胞系上统计pearson
#         if group.shape[0]>1:
#             smi_count +=1
#             metrics_dict_smi['Pearson'] += pearsonr(group['AUC'], group['AUC_pred'])[0]
#             metrics_dict_smi['RMSE'] += np.sqrt(mean_squared_error(group['AUC'], group['AUC_pred']))
#             metrics_dict_smi['R2'] += r2_score(group['AUC'], group['AUC_pred'])
#     for k in metrics_dict_cid.keys():
#         metrics_dict_cid[k] = metrics_dict_cid[k] / cid_count
#     for k in metrics_dict_smi.keys():
#         metrics_dict_smi[k] = metrics_dict_smi[k] / smi_count
#     return metrics_dict_cid, metrics_dict_smi

# def eval_metrics_for_group_list(df):
#     cid_count = 0
#     smi_count = 0
#     metrics_dict_cid = defaultdict(list)
#     metrics_dict_smi = defaultdict(list)
#     for cid, group in df.groupby('cid'):  #一个细胞系上所有分子统计pearson
#         if group.shape[0]>1:
#             cid_count += 1
#             metrics_dict_cid['cid'] += [cid]
#             metrics_dict_cid['Pearson'] += [pearsonr(group['AUC'], group['AUC_pred'])[0]]
#             metrics_dict_cid['RMSE'] += [np.sqrt(mean_squared_error(group['AUC'], group['AUC_pred']))]
#             metrics_dict_cid['R2'] += [r2_score(group['AUC'], group['AUC_pred'])]
#     for smi, group in df.groupby('smiles_idx'):  #一个分子在多个细胞系上统计pearson
#         if group.shape[0]>1:
#             smi_count +=1
#             metrics_dict_smi['cp_id'] += [smi]
#             metrics_dict_smi['Pearson'] += [pearsonr(group['AUC'], group['AUC_pred'])[0]]
#             metrics_dict_smi['RMSE'] += [np.sqrt(mean_squared_error(group['AUC'], group['AUC_pred']))]
#             metrics_dict_smi['R2'] += [r2_score(group['AUC'], group['AUC_pred'])]
# #     for k in metrics_dict_cid.keys():
# #         metrics_dict_cid[k] = metrics_dict_cid[k] / cid_count
# #     for k in metrics_dict_smi.keys():
# #         metrics_dict_smi[k] = metrics_dict_smi[k] / smi_count
#     return metrics_dict_cid, metrics_dict_smi

# def grid_search_for_RF_novel_chemical(data_pred, label, random_state, param_grid, scoring, groups, fold):
#     cv = GroupKFold(n_splits=fold)
#     clf = ensemble.RandomForestRegressor(random_state=random_state)
#     grid_search = GridSearchCV(clf, param_grid, scoring=scoring, n_jobs=6, verbose=2, cv=cv)
#     grid_search.fit(data_pred, label, groups=groups)
#     #传入所训练的数据
#     # RF_best = grid_search.best_estimator_
#     #best_estimator_是训练好的最优模型
#     # print('best score = ', grid_search.best_score_)
#     #输出训练的最优成绩
#     # print('best params:', grid_search.best_params_)
#     #输出最优参数
#     return grid_search.best_score_, grid_search.best_params_

# # def grid_search_for_RF_cell_blind(data_pred, label, random_state, param_grid, scoring, groups, fold):
# #     cv = GroupKFold(n_splits=fold)
# #     clf = ensemble.RandomForestRegressor(random_state=random_state)
# #     grid_search = GridSearchCV(clf, param_grid, scoring=scoring, n_jobs=6, verbose=2, cv=cv)
# #     grid_search.fit(data_pred, label, groups=groups)
# #     #传入所训练的数据
# #     # RF_best = grid_search.best_estimator_
# #     #best_estimator_是训练好的最优模型
# #     # print('best score = ', grid_search.best_score_)
# #     #输出训练的最优成绩
# #     # print('best params:', grid_search.best_params_)
# #     #输出最优参数
# #     return grid_search.best_score_, grid_search.best_params_


def pearson(actual, predict):
    score = pearsonr(actual, predict)[0]
    return score

def get_one_hot_cid(cid_set, cid_list):

    label_encoder = LabelEncoder()
    onehot_encoder = OneHotEncoder(sparse=False)

    label_encoder_fit = label_encoder.fit(cid_set)
    integer_encoded = label_encoder_fit.transform(cid_set)
    onehot_encoder_fit = onehot_encoder.fit(integer_encoded.reshape(-1, 1))
    cid_integer_encoded = label_encoder_fit.transform(cid_list)
    one_hot_encoded_cid = onehot_encoder_fit.transform(cid_integer_encoded.reshape(-1, 1))

    return one_hot_encoded_cid, onehot_encoder_fit, label_encoder_fit
```

```python
df_relation = pd.DataFrame(columns=['data', 'cid', 'value', 'relation_type'])
# random_seed = 1234
random_seed = 5678
# random_seed = 9999
# cell_ls = ['PC3']
# cell_ls = ['A375', 'MCF7', 'PC3', 'HT29']
for selected_cid in cell_ls:
    # print(selected_cid, df_LINCS_DR[df_LINCS_DR['cid'] == selected_cid].value_counts('label') )
    
#     for data_name, data_new in zip(['ECFP4+onehot cell', 'KPGT+onehot cell', 'DEG$_{real}$', 'SigGen', 
#                                # 'DLEPS+onehot cell', 'DeepCE', 'CIGER', 'KPGT+DEG$_{real}$'
#                                    ], 
#                                [ECFP_cid_array, KPGT_cid_array, DEG_array, DEG_pred_array, 
#                                 # DLEPS_pred_cid_array_new, DeepCE_pred_array_new, CIGER_pred_array_new, KPGT_DEG_real_array_new
#                                ]):
    for data_name, data_new in zip(['ECFP4', '$\mathit{\Delta}X$', 'TranSiGen'], [DEG_array, DEG_array, DEG_pred_array]):
        
        df_response = df_LINCS_DR[(df_LINCS_DR['cid'] == selected_cid) & (df_LINCS_DR['label'] == 1)]
        index_ls = df_response.index
        response_smiles = df_response['canonical_smiles'].tolist()
        response = data_new[index_ls,:]

        # index_ls = df_LINCS_DR[(df_LINCS_DR['cid'] == selected_cid) & (df_LINCS_DR['label'] == 0)].index
        df_no_response = df_LINCS_DR[(df_LINCS_DR['cid'] == selected_cid) & (df_LINCS_DR['label'] == 0)].sample(n=len(index_ls), 
                                                                                             random_state=random_seed)
        no_response_smiles = df_no_response['canonical_smiles'].tolist()
        index_ls = df_no_response.index
        no_response = data_new[index_ls,:]
        
        if data_name == 'ECFP4':
            response = []
            for smi in response_smiles:
                mol = Chem.MolFromSmiles(smi)
                ECFP = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
                response.append(ECFP)
            no_response = []
            for smi in no_response_smiles:
                mol = Chem.MolFromSmiles(smi)
                ECFP = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
                no_response.append(ECFP)
            
            pred_response_inter_pearson_ls = []
            for idx in range(len(response)):
                for idy in range(idx+1, len(response)):
                    sims = DataStructs.TanimotoSimilarity(response[idx], response[idy])
                    pred_response_inter_pearson_ls.append(sims)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, sims, 'Sensitive']

            pred_no_response_inter_pearson_ls = []
            for idx in range(len(no_response)):
                for idy in range(idx+1, len(no_response)):
                    sims = DataStructs.TanimotoSimilarity(no_response[idx], no_response[idy])
                    pred_no_response_inter_pearson_ls.append(sims)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, sims, 'Resistant']

            pred_response_no_response_pearson_ls = []
            for idx in range(len(response)):
                for idy in range(len(no_response)):
                    sims = DataStructs.TanimotoSimilarity(response[idx], no_response[idy])
                    pred_response_no_response_pearson_ls.append(sims)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, sims, 'Sensitive ~ Resistant']
        
        else:
            pred_response_inter_pearson_ls = []
            for idx in range(response.shape[0]):
                for idy in range(idx+1, response.shape[0]):
                    pear = pearsonr(response[idx, :], response[idy, :])[0]
                    pred_response_inter_pearson_ls.append(pear)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, pear, 'Sensitive']
            # print('pred_response_inter_pearson:', round(np.mean(pred_response_inter_pearson_ls),3))

            pred_no_response_inter_pearson_ls = []
            for idx in range(no_response.shape[0]):
                for idy in range(idx+1, no_response.shape[0]):
                    pear = pearsonr(no_response[idx, :], no_response[idy, :])[0]
                    pred_no_response_inter_pearson_ls.append(pear)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, pear, 'Resistant']
            # print('pred_no_response_inter_pearson:', round(np.mean(pred_no_response_inter_pearson_ls),3))

            pred_response_no_response_pearson_ls = []
            for idx in range(response.shape[0]):
                for idy in range(no_response.shape[0]):
                    pear = pearsonr(response[idx, :], no_response[idy, :])[0]
                    pred_response_no_response_pearson_ls.append(pear)
                    df_relation.loc[df_relation.shape[0]] = [data_name, selected_cid, pear, 'Sensitive ~ Resistant']
            print(data_name, selected_cid, np.mean(pred_response_inter_pearson_ls), 
                  np.mean(pred_no_response_inter_pearson_ls), 
                  np.mean(pred_response_no_response_pearson_ls) )     

df_relation.to_csv('/home/tongxiaochu/work/drug_perturbation/project/TranSiGen/results/5.Drug_response_prediction/relation_for_response.csv', index=False)
df_relation
```

```python
df_grid_search_results = pd.read_csv('../results/5.Drug_response_prediction/grid_search_param.csv')
df_data = pd.read_csv('../results/5.Drug_response_prediction/data.csv')

index_train = []
for _, row in df_data[df_data['split']=='train'].iterrows():
    index_train.append( df_LINCS[(df_LINCS['canonical_smiles']==row['canonical_smiles']) & (df_LINCS['cid']==row['cid'])].index[0] )
index_test = []
for _, row in df_data[df_data['split']=='test'].iterrows():
    index_test.append( df_LINCS[(df_LINCS['canonical_smiles']==row['canonical_smiles']) & (df_LINCS['cid']==row['cid'])].index[0] )

df_metrics = pd.DataFrame(columns=['Pearson', 'RMSE', 'R2', 'data'])
predict_result = defaultdict(list)
random_seed_ls  = [0, 1000, 10000, 100000, 1000000]
for data_name, data_sub in zip(['ECFP4+$\mathit{X}_1$', 'KPGT+$\mathit{X}_1$', '$\mathit{\Delta}X$', 'TranSiGen', 'DLEPS', 'DeepCE', 'CIGER'], 
                               [ECFP_cid_x1_array, KPGT_cid_x1_array, DEG_array, DEG_pred_array, DLEPS_pred_cid_array, DeepCE_pred_array, CIGER_pred_array]):
# for data_name, data_sub in zip(['TranSiGen', 'DLEPS'], 
#                                [DEG_pred_array, DLEPS_pred_cid_array]):
    print(data_name)
    best_params = df_grid_search_results[(df_grid_search_results['data']== data_name)].reset_index(drop=True).to_dict(orient='index')[0]
    print(best_params)
    
    metrics_dict = defaultdict(list)

    run_idx = 0

    
    for run_random_seed in random_seed_ls:
        clf = ensemble.RandomForestRegressor(max_depth=best_params['max_depth'], 
                                             n_estimators=best_params['n_estimators'], 
                                             max_features='auto', 
                                             criterion=best_params['criterion'], 
                                             oob_score = best_params['oob_score'],
                                             random_state=run_random_seed)

            
        data_train, data_test = data_sub[index_train], data_sub[index_test]
        label_train, label_test = np.array(df_data[df_data['split']=='train']['area_under_curve']), np.array(df_data[df_data['split']=='test']['area_under_curve'])
        smiles_train, smiles_test = np.array(df_data[df_data['split']=='train']['canonical_smiles']), np.array(df_data[df_data['split']=='test']['canonical_smiles'])
        smi_idx_train, smi_idx_test = np.array([smi2idx[smi] for smi in smiles_train]), np.array([smi2idx[smi] for smi in smiles_test])
        cid_train, cid_test = np.array(df_data[df_data['split']=='train']['cid']), np.array(df_data[df_data['split']=='test']['cid'])

        if data_name == 'DLEPS':
            train_index_new = np.where(data_train.any(axis=1))[0]
            print('train_index_new:', len(train_index_new))
            data_train = data_train[train_index_new,:]
            label_train = label_train[train_index_new]
            smi_idx_train = smi_idx_train[train_index_new]
            cid_train = cid_train[train_index_new]
            smi_idx_train_DLEPS = smi_idx_train
            
            test_index_new = np.where(data_test.any(axis=1))[0]
            print('test_index_new:', len(test_index_new))
            data_test = data_test[test_index_new,:]
            label_test= label_test[test_index_new]
            smi_idx_test = smi_idx_test[test_index_new]
            cid_test = cid_test[test_index_new]
            smi_idx_test_DLEPS = smi_idx_test
        
        print('TRAIN:', data_train.shape, 'TEST:', data_test.shape)
        print('train_smiles:', len(set(smi_idx_train)), 'test_smiles:', len(set(smi_idx_test)), 'common:', set(smi_idx_train) & set(smi_idx_test))
        print('train_cid:', len(set(cid_train)), 'test_cid:', len(set(cid_test)), 'common:', set(cid_train) & set(cid_test))

        clf = clf.fit(data_train, label_train)
        data_test_pred = clf.predict(data_test)
        if data_name == 'TranSiGen':
            predict_result['cp_id'] += [list(smi_idx_test)]
            predict_result['cid'] += [list(cid_test)]
            predict_result['label'] += [list(label_test)]
            predict_result['idx'] += [[run_idx] *len(list(label_test))]
        elif data_name == 'DLEPS':
            predict_result['DLEPS_cp_id'] += [list(smi_idx_test)]
            predict_result['DLEPS_cid'] += [list(cid_test)]
            predict_result['DLEPS_label'] += [list(label_test)]
            predict_result['DLEPS_idx'] += [[run_idx] *len(list(label_test))]
        
        predict_result[data_name] += [list(data_test_pred)]
        metrics_dict['Pearson'] += [pearsonr(label_test, data_test_pred)[0] ]
        metrics_dict['RMSE'] += [np.sqrt(mean_squared_error(label_test, data_test_pred))]
        metrics_dict['R2'] += [r2_score(label_test, data_test_pred)]
        metrics_dict['random_seed'] += [run_random_seed]

        df = pd.DataFrame(cid_test, columns=['cid'])
        df['cp_id'] = smi_idx_test
        df['AUC'] = label_test
        df['AUC_pred'] = data_test_pred
        print(df.shape)
        run_idx +=1
    df_metrics_tmp = pd.DataFrame.from_dict(metrics_dict)
    df_metrics_tmp.loc[:, 'data'] = data_name
    df_metrics = pd.concat([df_metrics, df_metrics_tmp])

with open('../results/5.Drug_response_prediction/prediction.pkl', 'wb') as f:
    pickle.dump(predict_result, f)
df_metrics.to_csv('../results/5.Drug_response_prediction/results.csv', index=False)

df_summary = df_metrics.groupby(['data']).agg(func=['mean', 'std'])
round(df_summary,3) 
```
