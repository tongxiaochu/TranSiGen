### DLESP

```python
df_result_DLEPS = pd.DataFrame(columns=['value', 'metrics', 'cid', 'seed'])

for random_seed in [343, 1234, 9876]:
    pred_dir = '/home/tongxiaochu/work/drug_perturbation/project/DLEPS-main/trained_model_same_split_DeepCE/' \
        'signature_test_10uM_7cells_common_DLEPS_pred_metrics_epoch200_seed{}_new.csv'.format(random_seed)
    df_test = pd.read_csv(pred_dir)
    
    for metrics in ['Pearson', 'RMSE', 'GSEA', 'Positive P@100', 'Negative P@100']:
        df = df_test[[metrics, 'cid']].rename(columns={metrics:'value'})
        df.loc[:, 'metrics'] = metrics
        df.loc[:, 'seed'] = random_seed
        df_result_DLEPS = pd.concat([df_result_DLEPS, df[['value', 'metrics', 'cid', 'seed']] ])
df_result_DLEPS.reset_index(inplace=True, drop=True)
df_result_DLEPS_cid = df_result_DLEPS.groupby(['metrics', 'cid']).mean().reset_index()
df_result_DLEPS_cid.loc[:, 'model'] = 'DLEPS'

df_result_DLEPS_run3 = df_result_DLEPS.groupby(['metrics', 'seed']).mean().reset_index()
df_result_DLEPS_run3.loc[:, 'model'] = 'DLEPS'
df_result_DLEPS_run3.groupby('metrics').mean()
```

```python
f_agg = ['mean', 'std']
df_summary = df_result_DLEPS_run3.groupby(['metrics']).agg(func=f_agg)
round(df_summary,3) #.columns
```

## CIGER

```python
df_result_CIGER = pd.DataFrame(columns=['value', 'metrics', 'cid', 'seed'])

for random_seed in [343, 1234, 9876]:
    pred_dir = '/home/tongxiaochu/work/drug_perturbation/project/CIGER-master/CIGER/saved_model/ciger_new/' 
    df_test = pd.read_csv(pred_dir + 'signature_test_10uM_7cells_pred_metrics_real_seed{}_new.csv'.format(random_seed) )
    df_test_reverse = pd.read_csv(pred_dir + 'signature_test_10uM_7cells_pred_metrics_real_reverse_seed{}_new.csv'.format(random_seed) )
    df_test['RMSE'] = (df_test['RMSE'] + df_test_reverse['RMSE'])/2
    df_test['Pearson'] = (df_test['Pearson'] + df_test_reverse['Pearson'])/2
    df_test['GSEA'] = (df_test['GSEA'] + df_test_reverse['GSEA'])/2
    df_test['Negative P@100'] = df_test_reverse['Negative P@100']
    for metrics in ['RMSE', 'Pearson', 'GSEA', 'Positive P@100', 'Negative P@100']:
        df = df_test[[metrics, 'cid']].rename(columns={metrics:'value'})
        df.loc[:, 'metrics'] = metrics
        df.loc[:, 'seed'] = random_seed
        df_result_CIGER = pd.concat([df_result_CIGER, df[['value', 'metrics', 'cid', 'seed']] ])
df_result_CIGER.reset_index(inplace=True, drop=True)
df_result_CIGER_cid = df_result_CIGER.groupby(['metrics', 'cid']).mean().reset_index()
df_result_CIGER_cid.loc[:, 'model'] = 'CIGER'

df_result_CIGER_run3 = df_result_CIGER.groupby(['metrics', 'seed']).mean().reset_index()
df_result_CIGER_run3.loc[:, 'model'] = 'CIGER'
df_result_CIGER_run3.groupby('metrics').mean()
```
