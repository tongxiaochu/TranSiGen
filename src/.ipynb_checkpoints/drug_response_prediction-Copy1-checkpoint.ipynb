```python
def eval_metrics_for_group(df):
    cid_count = 0
    smi_count = 0
    metrics_dict_cid = defaultdict(float)
    metrics_dict_smi = defaultdict(float)
    for cid, group in df.groupby('cid'):  #一个细胞系上所有分子统计pearson
        if group.shape[0]>1:
            cid_count += 1
            metrics_dict_cid['Pearson'] += pearsonr(group['AUC'], group['AUC_pred'])[0]
            metrics_dict_cid['RMSE'] += np.sqrt(mean_squared_error(group['AUC'], group['AUC_pred']))
            metrics_dict_cid['R2'] += r2_score(group['AUC'], group['AUC_pred'])
    for smi, group in df.groupby('smiles_idx'):  #一个分子在多个细胞系上统计pearson
        if group.shape[0]>1:
            smi_count +=1
            metrics_dict_smi['Pearson'] += pearsonr(group['AUC'], group['AUC_pred'])[0]
            metrics_dict_smi['RMSE'] += np.sqrt(mean_squared_error(group['AUC'], group['AUC_pred']))
            metrics_dict_smi['R2'] += r2_score(group['AUC'], group['AUC_pred'])
    for k in metrics_dict_cid.keys():
        metrics_dict_cid[k] = metrics_dict_cid[k] / cid_count
    for k in metrics_dict_smi.keys():
        metrics_dict_smi[k] = metrics_dict_smi[k] / smi_count
    return metrics_dict_cid, metrics_dict_smi

def eval_metrics_for_group_list(df):
    cid_count = 0
    smi_count = 0
    metrics_dict_cid = defaultdict(list)
    metrics_dict_smi = defaultdict(list)
    for cid, group in df.groupby('cid'):  #一个细胞系上所有分子统计pearson
        if group.shape[0]>1:
            cid_count += 1
            metrics_dict_cid['cid'] += [cid]
            metrics_dict_cid['Pearson'] += [pearsonr(group['AUC'], group['AUC_pred'])[0]]
            metrics_dict_cid['RMSE'] += [np.sqrt(mean_squared_error(group['AUC'], group['AUC_pred']))]
            metrics_dict_cid['R2'] += [r2_score(group['AUC'], group['AUC_pred'])]
    for smi, group in df.groupby('smiles_idx'):  #一个分子在多个细胞系上统计pearson
        if group.shape[0]>1:
            smi_count +=1
            metrics_dict_smi['cp_id'] += [smi]
            metrics_dict_smi['Pearson'] += [pearsonr(group['AUC'], group['AUC_pred'])[0]]
            metrics_dict_smi['RMSE'] += [np.sqrt(mean_squared_error(group['AUC'], group['AUC_pred']))]
            metrics_dict_smi['R2'] += [r2_score(group['AUC'], group['AUC_pred'])]
#     for k in metrics_dict_cid.keys():
#         metrics_dict_cid[k] = metrics_dict_cid[k] / cid_count
#     for k in metrics_dict_smi.keys():
#         metrics_dict_smi[k] = metrics_dict_smi[k] / smi_count
    return metrics_dict_cid, metrics_dict_smi

def grid_search_for_RF_novel_chemical(data_pred, label, random_state, param_grid, scoring, groups, fold):
    cv = GroupKFold(n_splits=fold)
    clf = ensemble.RandomForestRegressor(random_state=random_state)
    grid_search = GridSearchCV(clf, param_grid, scoring=scoring, n_jobs=6, verbose=2, cv=cv)
    grid_search.fit(data_pred, label, groups=groups)
    #传入所训练的数据
    # RF_best = grid_search.best_estimator_
    #best_estimator_是训练好的最优模型
    # print('best score = ', grid_search.best_score_)
    #输出训练的最优成绩
    # print('best params:', grid_search.best_params_)
    #输出最优参数
    return grid_search.best_score_, grid_search.best_params_

# def grid_search_for_RF_cell_blind(data_pred, label, random_state, param_grid, scoring, groups, fold):
#     cv = GroupKFold(n_splits=fold)
#     clf = ensemble.RandomForestRegressor(random_state=random_state)
#     grid_search = GridSearchCV(clf, param_grid, scoring=scoring, n_jobs=6, verbose=2, cv=cv)
#     grid_search.fit(data_pred, label, groups=groups)
#     #传入所训练的数据
#     # RF_best = grid_search.best_estimator_
#     #best_estimator_是训练好的最优模型
#     # print('best score = ', grid_search.best_score_)
#     #输出训练的最优成绩
#     # print('best params:', grid_search.best_params_)
#     #输出最优参数
#     return grid_search.best_score_, grid_search.best_params_


def pearson(actual, predict):
    score = pearsonr(actual, predict)[0]
    return score

def get_one_hot_cid(cid_set, cid_list):

    # values = ['PC3', 'MCF7', 'A375', 'HA1E', 'VCAP', 'A549', 'HEPG2', 'HT29']
    # values = ['PC3', 'MCF7', 'A375', 'HA1E', 'A549', 'HEPG2', 'HT29']

    label_encoder = LabelEncoder()
    onehot_encoder = OneHotEncoder(sparse=False)

    label_encoder_fit = label_encoder.fit(cid_set)
    integer_encoded = label_encoder_fit.transform(cid_set)
    onehot_encoder_fit = onehot_encoder.fit(integer_encoded.reshape(-1, 1))
    cid_integer_encoded = label_encoder_fit.transform(cid_list)
    # print(drug_integer_encoded)
    one_hot_encoded_cid = onehot_encoder_fit.transform(cid_integer_encoded.reshape(-1, 1))
    # print(one_hot_df)
    return one_hot_encoded_cid, onehot_encoder_fit, label_encoder_fit
```
